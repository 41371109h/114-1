```python
import datetime
from google.colab import auth
import gspread
from google.auth import default

# --- Google Sheet 認證 ---
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# 連接 Google Sheet（改成你的表單網址）
sheet_url = 'https://docs.google.com/spreadsheets/d/1N0nOCQnyxrJqLYYOr6aL2rvluNmZvd0plJYzbuwJMu4/edit?usp=sharing'
gsheets = gc.open_by_url(sheet_url)
worksheet = gsheets.worksheet('工作表1')   # 改成你的工作表名稱

def record_expenses_loop():
    """
    持續輸入消費資訊，直到輸入 'q' 停止，並將紀錄寫入 Google Sheet。
    """
    while True:
        print("\n--- 新增一筆消費紀錄 (輸入 q 離開) ---")

        # 取得當前的日期
        date_str = datetime.datetime.now().strftime("%Y-%m-%d")

        # 讓使用者輸入品項
        item = input("請輸入品項：")
        if item.lower() == "q":
            print("已結束輸入。")
            break

        # 讓使用者輸入金額
        amount = input("請輸入金額：")
        if amount.lower() == "q":
            print("已結束輸入。")
            break

        # 讓使用者輸入類別
        category = input("請輸入類別：")
        if category.lower() == "q":
            print("已結束輸入。")
            break

        # 新增一列資料
        new_row = [date_str, item, amount, category]

        try:
            worksheet.append_row(new_row)
            print("✅ 已成功將紀錄新增到 Google Sheet！")
        except Exception as e:
            print(f"寫入 Google Sheet 時發生錯誤：{e}")

# 測試執行
if __name__ == "__main__":
    record_expenses_loop()
```
